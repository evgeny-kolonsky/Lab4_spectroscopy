# Introduction

Spectroscopy is the basic method of studying energy levels.  In this laboratory, we will use spectroscopic methods to study hydrogen energy levels.  When an atom moves from one energy level to another lower than it, it emits a photon whose energy, $h\nu$, is equal to the difference between these two energy levels.  Since atomic energy levels are discrete, the radiated light is composed of discrete wavelengths.  

The function of the spectrometer is to split the light into the bitter wavelengths that turn it off and measure them.  Knowing these wavelengths, it is possible to build the energy level scale of the atom.

Spectroscopic measurements are among the most accurate.  Even in our own laboratory, we expect inaccuracies of less than one percent.  Below we describe two different spectroscopic methods using a lattice and using a prism.

## Difraction lattice

image.png

An indirect lattice is a dense network of cracks.  When a monochromatic light beam falls on an indirect lattice, then a maxim of light at certain diversion angles is obtained as a result of interference.  

If the light falls perpendicular to the lattice,then the condition for these charms is:
$$
d \sin \vartheta = n \lambda, \quad n = 0, \pm 1, \pm 2, ..
\tag{1}
$$
$\lambda$ is the wavelength and $d$ the distance betweenthe cracks. $n$ is called the order of diffraction.  Light is separated into different wavelengths because the angles of the maxima depend on the wavelength (except for the order of 0).  Figure 3 depicts the separation of wavelengths and several different orders.

Interference also exists when light does not fall perpendicular to the lattice, but then the condition for the angle of diversion of maximum light is not given by the formula (1), but by the angle between the direction of the offending light and the lattice.  It can be shown that the shift angle of a certain maximum is minimal when the lattice stands symmetrically with respect to the incoming beam and the (distracted) outgoing beam.  This situation is illustrated in Figure 4.  A beam of lighthits the lattice at an angle $\alpha$ perpendicular to the lattice.The angle of diversionis:
$$
 \alpha = {1\over 2} \vartheta_{min}
 \tag{2}
$$

A condition for a minimum shif tangle for an $n$-order maximum is:
$$
2 d \sin (\vartheta_{min}/2) = n \lambda, \quad n = 0, \pm 1, \pm 2, ..
\tag{3}
$$

Measuring the minimum angle of diversion that eliminates the need to place the lattice perpendicular to the light beam, and in principle its measurement is more accurate.

## Energy levels in the hydrogen atom

The $n$-th level in the hydrogen atom is given by:
$$
E_n = R_y {1\over n^2}, \quad n = 1,2,3,4..
$$
Where $n$ is called *first quantum number* and $R_y$ is *Ridberg's constant*
$ R_y = 13.6eV $.
When the hydrogen atom is excited, for example by colliding with a decomposition tube, the electrons are excited to higher energy levels.  Then, descending back to the lower levels, they emit light with an energy equal to the energy difference between the levels:
$$
h\nu = R_y \left( {1 \over n_i^2} - {1 \over n_f^2}\right)
\tag{5}
$$
